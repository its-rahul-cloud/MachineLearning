```python
p0=data.Purchase.min()
p100=data.Purchase.max()
q1=data.Purchase.quantile(0.25)
q2=data.Purchase.quantile(0.5)
q3=data.Purchase.quantile(0.75)
iqr=q3-q1
```

```python
#lower cutoff
lc=q1-1.5*iqr
lc
```

```python
#upper cutoff
uc=q3+1.5*iqr
uc
```

```python
data.Purchase.plot(kind='box')
```

```python
data.Purchase.clip(upper=uc)
```

```python
data.Purchase.clip(upper=uc,inplace=True)
data.Purchase.plot(kind='box')
```
